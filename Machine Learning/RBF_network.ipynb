```python
def kmeans(X, k):
    clusters = np.random.choice(np.squeeze(X), size=k)
    clusters_prev, stds = clusters.copy(), np.zeros(k)
    converged = False

    while not converged:
        distances = np.squeeze(np.abs(X[:, np.newaxis] - clusters[np.newaxis, :]))
        cluster_closed = np.argmin(distances, axis=1)
        for i in range(k):
            pts = X[cluster_closed == i]
            if len(pts) > 0:
                clusters[i] = np.mean(pts, axis=0)

        converged = np.linalg.norm(clusters-clusters_prev) < 1e-6
        clusters_prev = clusters.copy()

    distances = np.squeeze(np.abs(X[:, np.newaxis] - clusters[np.newaxis, :]))
    cluster_closed = np.argmin(distances, axis=1)

    clusters_pts = []
    for i in range(k):
        pts = X[cluster_closed == i]
        if len(pts) < 2:
            clusters_pts.append(i)
            continue
        
        else:
            stds[i] = np.std(X[cluster_closed == i])

    if len(clusters_pts) > 0:
        pts_avg = []
        for i in range(k):
            if i not in clusters_pts:
                pts_avg.append(X[cluster_closed == i])
        pts_avg = np.concatenate(pts_avg).ravel()
        stds[clusters_pts] = np.mean(np.std(pts_avg))

    return clusters, stds
```

```python
def rbf(x, c, s): # radial (or Gaussian) kernels
    return np.exp(-1*(x-c)**2 / (2*s**2))
```

```python
class CLF_RBFN(object):
    def __init__(self, k, sigma=1.0):
        self.k = k # the number of bases
        self.sigma = sigma
        self._centers = None
        self.weights = None

    def basis_function(self, center, data):
        return np.exp(- self.sigma*np.linalg.norm(center - data)**2)

    def interpolation_matrix(self, X):
        gauss = np.zeros((len(X), self.k))
        for data_arg, data in enumerate(X):
            for center_arg, center in enumerate(self._centers):
                gauss[data_arg, center_arg] = self.basis_function(center, data)
                
        return gauss

    def centers(self, X):
        random_args = np.random.choice(len(X), self.k)
        return X[random_args]

    def fit(self, X, Y):
        self._centers = self.centers(X)
        gauss = self.interpolation_matrix(X)
        self.weights = np.dot(np.linalg.pinv(gauss), Y)

    def predict(self, X):
        gauss = self.interpolation_matrix(X)
        
        return np.dot(gauss, self.weights)
```

```python
class REG_RBFN(object):
    def __init__(self, k=2, lr=0.01, epochs=100, rbf=rbf, Stds=True):
        self.k = k # the number of bases
        self.lr = lr
        self.epochs = epochs
        self.rbf = rbf
        self._Stds = Stds

        self.w = np.random.randn(k)
        self.b = np.random.randn(1)
        
    def fit(self, X, y):
        if self._Stds:
            self.centers, self.stds = kmeans(X, self.k)
        else:
            self.centers, __ = kmeans(X, self.k)
            self.stds = np.repeat(max([np.abs(
                c1 - c2) for c1 in self.centers for c2 in self.centers]
                                     ) / np.sqrt(2 * self.k), self.k)

        for epoch in range(self.epochs):
            for i in range(X.shape[0]):
                a = np.array([self.rbf(X[i], c, s) for c, s, in zip(
                    self.centers, self.stds)])
                F = a.T.dot(self.w) + self.b

                error = - (y[i] - F).flatten()

                self.w = self.w - self.lr * a * error
                self.b = self.b - self.lr * error
                
    def predict(self, X):
        y_pred = []
        for i in range(X.shape[0]):
            a = np.array([self.rbf(X[i], c, s) for c, s, in zip(
                self.centers, self.stds)])
            F = a.T.dot(self.w) + self.b
            
            y_pred.append(F)
            
        return np.array(y_pred)
```

```python
# For circle in the square data
from sklearn.metrics import accuracy_score
```

```python
train_x, train_y = np.array(clf_train1['x']), np.array(clf_train1['y'])
train_z = np.array(clf_train1['label'])

train_features = np.asarray(list(zip(train_x.flatten(), train_y.flatten())))

clf_rbfn = CLF_RBFN(sigma=1., k=10)
clf_rbfn.fit(train_features, train_z)

test_x, test_y = np.array(clf_test['x']), np.array(clf_test['y'])
test_z = np.array(clf_test['label'])
test_z = np.where(test_z > 0, 0, 1)

test_features = np.asarray(list(zip(test_x.flatten(), test_y.flatten())))
z_pred = clf_rbfn.predict(test_features)
z_converted = np.where(z_pred > np.mean(z_pred), 0, 1)

plt.figure(figsize=(10, 10))
plt.scatter(test_features[:, 0], test_features[:, 1], c=z_converted)
plt.title('Circle in the square with train 1 and test data', fontsize=20)
plt.text(0.35, 0.5, 'Accuracy: {:.2f}%'.format(
    accuracy_score(test_z, z_converted)*100), fontsize=20)
plt.show()

print(np.mean(z_converted), np.var(z_converted))
```
