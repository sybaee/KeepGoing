```python
for word in temp:
    print(word)
    category = input()
    
    if category is '1' or category is '2' or category is '3':
        dict_category[int(category)].append(word)
    
    elif category is '4':
        vague.append(word)
    
    elif category is '5':
        later.append(word)
    
    elif category is '9':
        print(edited)
        print(vague)
        print(later)
        print(dict_category)
        
    elif category is '0':
        continue
    
    else:
        edited.append(word)
```
