```python
# read useful data list
with open(fig_path, "r") as f:
    keys = f.read().split('\n')
    
del keys[-1] # delete newline
df_useful = {key.split(', ')[0]: {} for key in keys}
category_item = [[key.split(', ')[0], key.split(', ')[1]] for key in keys]

for pair in category_item:
    print(pair)
    category, item = pair[0], pair[1]
    df_useful[category][item] = dict_item_dfs[category][item].copy(deep=True)
```

```python
def get_anomaly(df, standard, up_diff, down_diff, max_count):
    count = 0
    x1 = df['DEMAND'][0]

    anomaly = []
    anomaly.append({'DATE': df['DATE'][0], 'DEMAND': x1})

    for i in range(len(df)):
        try:
            count += 1
            x2 = df['DEMAND'][i]
            
            diff = x2 - x1
            
            # detect anomaly to point out the soaring range
            if count < max_count:
                if x1 < standard:
                    if (diff > 0 and diff >= up_diff[0]) or (diff < 0 and abs(diff) >= down_diff[0]):
                        anomaly.append({'DATE': df['DATE'][i], 'DEMAND': x2})
                        x1 = x2

                else:
                    if (diff > 0 and diff >= up_diff[1]) or (diff < 0 and abs(diff) <= down_diff[1]):
                        anomaly.append({'DATE': df['DATE'][i], 'DEMAND': x2})
                        x1 = x2
            
            else:
                if (diff > 0 and diff <= up_diff[0]) or (diff < 0 and abs(diff) >= down_diff[1]):
                    anomaly.append({'DATE': df['DATE'][i], 'DEMAND': x2})
                    x1 = x2
                    count = 0
                                     
        except IndexError:
            return print('IndexError occured')
        
    return anomaly
```

```python
def construct_anomaly_df(df, anomaly):
    df_anomaly = pd.DataFrame(columns=['DATE', 'DEMAND'])
    df_anomaly['DATE'] = df['DATE']
    
    for values in anomaly:       
        df_anomaly['DEMAND'].mask(
            df_anomaly['DATE'] == values['DATE'], values['DEMAND'], inplace=True)
    
    return df_anomaly
```

```python
def interpoloate_df(df):
    df.index = pd.to_datetime(df['DATE'])
    del df['DATE']
    
    df['DEMAND'] = pd.to_numeric(df['DEMAND'])
    
    # return dataframe with fully filled NaN rows
    return df.resample('1d').interpolate('time').reset_index()
```

```python
mean_tf = False
save_tf = False
save_csv_path = './data_smoothing/csv'
save_fig_path = './data_smoothing/figure'

params = [
    [4, (3, 1), (5, 2), 10],
    [4, (3, 1), (5, 2), 20],
    [4, (3, 1), (5, 2), 30],
    [5, (3, 1), (5, 2), 10],
    [5, (3, 1), (5, 2), 20],
    [5, (3, 1), (5, 2), 30],
    [6, (3, 1), (5, 2), 10],
    [6, (3, 1), (5, 2), 20],
    [6, (3, 1), (5, 2), 30],
    
    [4, (3, 1), (5, 3), 10],
    [4, (3, 1), (5, 3), 20],
    [4, (3, 1), (5, 3), 30],
    [5, (3, 1), (5, 3), 10],
    [5, (3, 1), (5, 3), 20],
    [5, (3, 1), (5, 3), 30],
    [6, (3, 1), (5, 3), 10],
    [6, (3, 1), (5, 3), 20],
    [6, (3, 1), (5, 3), 30]
]

for param in params:
    standard, up_diff, down_diff, max_count = param[0], param[1], param[2], param[3]
    
    for category in df_useful.keys():
        for item, df in df_useful[category].items():
            if mean_tf: # write filled-with-mean dataframe file
                df_mean = interpoloate_df(df.copy(deep=True))
                df_mean.to_csv('{}/df_mean_{}_{}.csv'.format(
                    save_csv_path, category, item), index=False, header=True)
            
            anomaly = get_anomaly(df, standard, up_diff, down_diff, max_count)
            df_anomaly = interpoloate_df(construct_anomaly_df(df, anomaly))

            fig, ax = plt.subplots(figsize=(16, 8))

            df.reset_index().plot(x='index', y='DEMAND', kind='scatter', ax=ax)
            df_anomaly.reset_index().plot(x='index', y='DEMAND', kind='line', color='red', ax=ax)

            fig.suptitle(item + ' in ' + category, fontsize=15, y=0.92)
            plt.xlabel('date range', fontsize=12)
            plt.ylabel('item quantity', fontsize=12)
            plt.legend(['DEMAND', 'DEMAND'])

            if save_tf: # write anomaly dataframe file and figure
                df_anomaly.to_csv('{}/df_{}_{}_sd_{}_up_{}_{}_down_{}_{}_cnt_{}.csv'.format(
                    save_csv_path, category, item, 
                    standard, up_diff[0], up_diff[1], down_diff[0], down_diff[1], max_count
                ), index=False, header=True)
                plt.savefig('{}/fig_{}_{}_sd_{}_up_{}_{}_down_{}_{}_cnt_{}.png'.format(
                    save_fig_path, category, item, 
                    standard, up_diff[0], up_diff[1], down_diff[0], down_diff[1], max_count))
            
            else:
                plt.show()
                import pdb; pdb.set_trace() # type 'c + ENTER' to continue to draw graph
    
    mean_tf = False
```
