```python
dict_file = {
    'composition': [],
    'item': [],
    'color': [],
    'brand': [],
    'category': [],
    'etc': []
}
```

```python
for i in range(1, len(df.description)):
    print("=========================================================   paragraph " + str(i))
    para = df.description[i]
    print(para)
    para = para.split("\n")
    print("=========================================================")
    for line in para:
        print(line)
        print("---------------------------------------------------------")
        
        while True:
            print("\n***** keyword")
            keyword = input()
            if keyword == "":
                break
                
            print("\n>>> choice")
            print("1: composition, 2: item, 3: color, 4: brand, 5: category, ENTER: etc")
            choice = input()
            print()
            if choice == '1':
                dict_file['composition'].append(keyword)
            elif choice == '2':
                dict_file['item'].append(keyword)
            elif choice == '3':
                dict_file['color'].append(keyword)
            elif choice == '4':
                dict_file['brand'].append(keyword)
            elif choice == '5':
                dict_file['category'].append(keyword)
                
            elif choice == '8':
                for key in dict_file.keys():
                    print(dict_file[key])
            elif choice == '9':
                for key in dict_file.keys():
                    words = list(set(dict_file[key]))
                    file = './dict_file/' + key + '.txt'
                    with open(file, 'a') as f:
                        for word in words:
                            f.write(str(word) + '\n')

            else:
                dict_file['etc'].append(keyword)
        print("---------------------------------------------------------")
```
